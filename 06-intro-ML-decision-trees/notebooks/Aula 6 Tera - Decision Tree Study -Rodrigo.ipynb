## Iris case
>The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis. [4]




| Elements | Taxonomy   |
|------|------|
|![Taxonomy hierarchy](../data/taxonomyHierarchy.png) [5]|![Iris elements](https://evolution.berkeley.edu/evolibrary/images/interviews/flower_diagram.gif "Iris elements") [1]|

Feature: Caracteristica do elemento de estudo.

Target: Variavel que se busca predizer.

```python
#--Library
import os
import subprocess
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
#--Load iris dataset
iris = datasets.load_iris()
print(iris.DESCR)
```

```python
# Data from iris
print(dict(iris))
```

```python
help(iris)
```

```python
help(np.c_)
```

```python
# Preparando o dataset para o dataframe
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
df = pd.DataFrame(data=iris['data'],columns=features)

# O preditor funciona melhor quando o target é int
df['Target'] = iris['target'].astype(int)

# Test rows list
rows_test = [49,99,149]

# DataFrame de Treino
df_train = df.drop(df.index[rows_test])

# Dataframe de Test
df_test = df.iloc[rows_test]

# Visualizar parte do dataframe
df.iloc[40:60]
```

```python
#--Model data
Y = df_train['Target']
X = df_train[features]

dt = DecisionTreeClassifier(min_samples_split=20, random_state=99) #parameters to understand and modify
dt.fit(X, Y) #Train model
```

```python
#--Take a look at X (features) and y (target)
print("X:\n",X.head())
print("Y:\n",Y.head())
```

```python
#--Predict target
#dt.predict([[3,5,3,0]])
#dt.predict([[3,5,1,0]])
dt.predict(df_test[df_test['Target'] == 0][features])
```

```python
#--Check tree accuracy
dt.score(X,Y)
```
