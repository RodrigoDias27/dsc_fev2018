## Adicionando mais curvas na Janela

### Variáveis

```python
""" definindo as variáveis """
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x * 0.333334
```

### Comando único

```python
""" construindo o plot """
plt.figure(figsize=(12, 8))

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("senóide")

plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.grid(True)
```

```python
""" construindo o plot """
plt.plot(
    x, y1, "r-",
    x, y2, "g--",
    x, y3, "b:"
)
```

```python
""" inserindo a legenda; descomente outras linhas para variar a posição da legenda """
local = "best"
# local = "upper right"
# local = "upper left"
# local = "lower left"
# local = "lower right"
# local = "right"
# local = "center left"
# local = "center right"
# local = "lower center"
# local = "upper center"
# local = "center"

plt.legend(["seno", "cosseno", "reta"], loc=local)
```

```python
plt.show()
```

### Método Aditivo

```python
""" construindo o plot """
plt.figure(figsize=(12, 8))

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("senóide")

plt.xlim(-np.pi, np.pi)
plt.ylim(-1.2, +1.2)

plt.grid(True)
```

```python
""" construindo o plot """
plt.plot(x, y1, color="red", linewidth=2, linestyle="-")
plt.plot(x, y2, color="green", linewidth=4, linestyle="--")
plt.plot(x, y3, color="blue", linewidth=6, linestyle=":")
```

```python
""" inserindo a legenda; descomente outras linhas para variar a posição da legenda """
# por padrão, loc = "best"
plt.legend(["seno", "cosseno", "reta"])
```

```python
plt.show()
```

## Salvando em arquivos

#### Variáveis

```python
""" definindo as variáveis """
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x * 0.333334
y4 = np.tan(x)
y5 = np.cosh(x)
```

### Salvando em PDF e PNG

```python
""" construindo o plot """
plt.figure(figsize=(12, 8))

plt.xlabel("eixo X")
plt.ylabel("eixo Y")
plt.title("senóide")

plt.xlim(-np.pi, np.pi)
plt.ylim(-2, +5)

plt.grid(True)

plt.plot(x, y1, color="red", linewidth=1, linestyle="-")
plt.plot(x, y2, color="green", linewidth=2, linestyle="--")
plt.plot(x, y3, color="blue", linewidth=3, linestyle=":")
plt.plot(x, y4, color="darkorange", linewidth=4, linestyle="--")
plt.plot(x, y5, color="k", linewidth=5, linestyle="-")


plt.legend(["seno", "cosseno", "reta", "tangente", "cosseno hiperbólico"])
```

```python
""" salvando PDF """
plt.savefig("senoide.pdf")
```

```python
""" salvando PNG """
plt.savefig("senoide.png")
```

```python
""" mostrando apenas após salvar """
plt.show()
```

# Seaborn

## Importando o Seaborn

```python
import seaborn as sns
```

### Variáveis

```python
""" definindo as variáveis """
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x * 0.333334
```

## Rodando o mesmo experimento com o Seaborn importado

```python
""" encapsulando o plot """
def plot_curves():
    plt.figure(figsize=(12, 8))
    plt.plot(x, y1, color="red")
    plt.plot(x, y2, color="green")
    plt.plot(x, y3, color="blue")
    plt.legend(["seno", "cosseno", "reta"])
    plt.show()
```

```python
""" plot sem SNS """
plot_curves()
```

```python
""" plot com SNS """
sns.set()
plot_curves()
```
